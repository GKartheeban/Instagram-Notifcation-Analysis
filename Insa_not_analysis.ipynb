```python
%%sql
create database notification_db;
```

```python
%%sql
use notification_db;
```

```python
%%sql
CREATE TABLE notifications (
    id INT AUTO_INCREMENT PRIMARY KEY,
    title VARCHAR(255) NOT NULL,
    body TEXT NOT NULL,
    type ENUM('foreground', 'background') NOT NULL,
    source VARCHAR(50) DEFAULT 'Instagram',  -- New column for source
    received_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
```

```python
%%sql
ALTER TABLE notifications ADD UNIQUE KEY unique_notification (title, body(255), type);
```

```python
%%sql
select * from notifications
```

```python
%%sql
select substring_index(body, ' ', 1) as username,
count(*) as count
from notifications
group by username
order by count desc
limit 5;
```

```python
%%sql
select type, count(*) as count
from notifications
group by type
```
